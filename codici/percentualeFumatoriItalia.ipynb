```python
import plotly.graph_objects as go

# Dati
val_2023 = 20.5
val_2024 = 25.4
delta = round(val_2024 - val_2023, 1)

# Colori per tutto
COLOR_2024 = "#5DADE2"    # Blu cielo
COLOR_2023 = "#1a237e"    # Blu Profondo
COLOR_TEXT = "#455a64"
COLOR_DANGER = "#e74c3c"  # Rosso per l'aumento
COLOR_GOOD = "#2ecc71"    # Verde per la diminuzione
FONT_FAMILY = "Arial, sans-serif"

# Creazione della figura
fig = go.Figure()

# Arco ESTERNO (2024)
fig.add_trace(go.Indicator(
    mode = "gauge",
    value = val_2024,
    domain = {'x': [0, 1], 'y': [0, 1]},
    gauge = {
        'shape': "angular",
        'axis': {
            'range': [0, 100],
            'tickmode': 'linear',
            'dtick': 10,
            'tickwidth': 2,
            'ticklen': 10,
            'tickcolor': COLOR_TEXT,
            'visible': True,
            'showticklabels': True,
            'tickfont': {'color': COLOR_TEXT, 'size': 12}
        },
        'bar': {'color': COLOR_2024, 'thickness': 0.20},
        'bgcolor': "rgba(0,0,0,0.05)",
        'borderwidth': 0,
        # Questo è per la linea che attraversa i grafici
        'threshold': {
            'line': {'color': COLOR_2023, 'width': 2},
            'thickness': 1,
            'value': val_2023
        }
    }
))

# Arco INTERNO (2023)
fig.add_trace(go.Indicator(
    mode = "gauge",
    value = val_2023,
    domain = {'x': [0.15, 0.85], 'y': [0.05, 0.75]},
    gauge = {
        'shape': "angular",
        'axis': {
            'range': [0, 100],
            'tickmode': 'linear',
            'dtick': 10,
            'visible': True,
            'showticklabels': False
        },
        'bar': {'color': COLOR_2023, 'thickness': 0.18},
        'bgcolor': "rgba(0,0,0,0.05)",
        'borderwidth': 0,
        # Questo è per la liea che attraversa i grafici
        'threshold': {
            'line': {'color': COLOR_2023, 'width': 2},
            'thickness': 1,
            'value': val_2023
        }
    }
))

# Numeri centrali
fig.add_annotation(
    x=0.5, y=0.25,
    text=(
        f"<span style='color:{COLOR_2024}; font-size:45px; font-weight:bold;'>{val_2024}</span> "
        f"<span style='color:gray; font-size:30px;'>-</span> "
        f"<span style='color:{COLOR_2023}; font-size:35px; font-weight:bold;'>{val_2023}</span>"
    ),
    showarrow=False,
    xref="paper", yref="paper"
)

# Indicatore di crescita/decrescita
symbol = "▲" if delta > 0 else "▼"
delta_color = COLOR_DANGER if delta > 0 else COLOR_GOOD # Rosso se aumenta, Verde se cala

fig.add_annotation(
    x=0.5, y=0.18,
    text=f"<b style='color:{delta_color}; font-size:28px;'>{symbol} +{abs(delta)}%</b>",
    showarrow=False,
    xref="paper", yref="paper"
)

# TITOLO E SOTTOTITOLO
fig.add_annotation(
    x=0.5, y=1.20,
    text="<b style='color:#2E4053; font-size:32px;'>Percentuale di Fumatori in Italia</b>",
    showarrow=False,
    xref="paper", yref="paper"
)

fig.add_annotation(
    x=0.5, y=1.10,
    text="<span style='color:#7F8C8D; font-size:16px;'>Confronto 2023 vs 2024 • Campioni di 10 milioni di abitanti</span>",
    showarrow=False,
    xref="paper", yref="paper"
)

# Legenda anni ~meh~
fig.add_annotation(
    x=1, y=1,
    text=(
        f"<span style='color:{COLOR_2024}; font-size:20px;'>■</span> <b>2024</b><br>"
        f"<span style='color:{COLOR_2023}; font-size:20px;'>■</span> <b>2023</b>"
    ),
    showarrow=False,
    align="left",
    xref="paper", yref="paper",
    font=dict(size=14, color=COLOR_TEXT),
    bgcolor="rgba(255,255,255,0.7)",
    bordercolor="#D5DBDB",
    borderwidth=1,
    borderpad=10
)

# Layout finale
fig.update_layout(
    paper_bgcolor = "white",
    plot_bgcolor = "white",
    margin=dict(l=60, r=60, t=140, b=60),
    width=900,
    height=750,
    font={'family': FONT_FAMILY}
)

fig.show()
```
